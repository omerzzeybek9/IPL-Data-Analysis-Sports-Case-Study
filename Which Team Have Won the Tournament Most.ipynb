2 - Which Team Have Won The Most?

```python
df["Season"] = pd.to_datetime(df["date"]).dt.year
df["Season"].unique()
```

```python
#Winners of the tournament each season (last row tells you the winner)
df[df["Season"] == 2018]["winner"].tail(1).values[0]
```

```python
winners = {}
for year in sorted(df["Season"].unique()):
    current_winner = df[df["Season"] == year]
    winners[year] = current_winner["winner"].tail(1).values[0]
```

```python
winners
```

```python
#Time to count who won the tournament most
from collections import Counter

Counter(winners.values())
```

```python
#Plot the data
import matplotlib.pyplot as plt


winners = Counter({'Mumbai Indians': 5,
                   'Chennai Super Kings': 3,
                   'Kolkata Knight Riders': 2,
                   'Rajasthan Royals': 1,
                   'Deccan Chargers': 1,
                   'Sunrisers Hyderabad': 1})

winners_dict = dict(winners)

teams = list(winners_dict.keys())
counts = list(winners_dict.values())

plt.figure(figsize=(10, 6))
plt.bar(teams, counts, color='skyblue')
plt.xlabel('Teams')
plt.ylabel('Number of Tournaments Won')
plt.title('Top IPL Tournament Winners')
plt.xticks(rotation=45)
plt.show()
```
